```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from keras.utils import np_utils
```

```python
#Importing dataset
dataset = pd.read_csv('train.csv')
```

```python
#Splitting inputs and outputs
X = dataset.iloc[:,1:785].values.astype(float)
y = dataset.iloc[:,0].values
```

```python
#Reshaping input array
X = np.array(X).reshape(y.shape[0], 28, 28, 1).astype(float)
X = X/255.0
```

```python
#Encoding dependent variable
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit(y)
y = encoder.transform(y)
# convert integers to dummy variables (i.e. one hot encoded)
y = np_utils.to_categorical(y)
```

```python
# Importing the Keras libraries and packages for CNN
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers.advanced_activations import LeakyReLU as LR
```

```python
# Initialising the CNN
classifier = Sequential()

# Step 1 - Convolution
classifier.add(Convolution2D(64, 3, 3, input_shape = (28, 28, 1), activation = LR(alpha = 0.3)))

# Step 2 - Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Adding a second convolutional layer
classifier.add(Convolution2D(96, 3, 3, activation = LR(alpha = 0.3)))
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Adding a THird convolutional layer
classifier.add(Convolution2D(128, 3, 3, activation = LR(alpha = 0.3)))
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Step 3 - Flattening
classifier.add(Flatten())

# Step 4 - Full connection
classifier.add(Dense(output_dim = 256, activation = LR(alpha = 0.3)))
classifier.add(Dense(output_dim = 10, activation = 'softmax'))
```

```python
# Compiling the CNN and Training
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
classifier.fit(X, y, batch_size = 128, nb_epoch = 20)
```

```python
#Loading Test data
submission_set = pd.read_csv('test.csv')

#Creation of ImageId
id_ = [i for i in xrange(28001)]
id_ = id_[1:]
test_id = np.asarray(id_)

#Preparing Dataset
X_submission = submission_set.values.astype(float)
#Reshaping input array
X_submission = np.array(X_submission).reshape(X_submission.shape[0], 28, 28, 1).astype(float)
X_submission = X_submission/255.0
```

```python
#Predicting the Result
y = classifier.predict(X_submission)
y  = (y > 0.5)
y = y.argmax(1)
```

```python
#Writing to CSV
CSVFile = np.column_stack((test_id.T,y.T))
outdata = pd.DataFrame(CSVFile, columns = ['ImageId', 'Label'])
outdata.to_csv('output_cnn_ipynb.csv', index = None)
```
